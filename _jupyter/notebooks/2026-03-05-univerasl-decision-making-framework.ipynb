## Introduction
Decision-making — whether in games, business, or everyday life — is often treated as a process of choosing between discrete, isolated actions. In Chess or Go, we imagine moving a single piece toward a goal. In business, we think in terms of tackling one project or opportunity at a time. While this simplification is useful, it obscures the deeper structure that governs strategic success: the interplay between an agent’s current state, the options available, and the evolution of those states over time.

At the heart of every decision lies **position**: the agent’s state vector $P = [p_1, p_2, \dots, p_n]$, which encodes the resources, skills, knowledge, and context relevant to that agent. In games, position might be the arrangement of pieces on a board; in business, it could combine revenue, security, and skill; in life more broadly, it spans finances, health, social networks, and knowledge. Crucially, a high-quality position is not defined solely by immediate advantage. It is one that **maximizes future optionality**, increasing the number and quality of paths to success, while **minimizing exposure to threats**, the pathways that lead to failure. Unlike classical “system-centered” game trees, which enumerate all players’ possible moves, this framework focuses on the **agent-centered game tree**, projecting the global system onto the agent’s accessible and controllable state space.

Complementing position is the concept of a **move**. Traditional thinking treats moves as atomic actions — a pawn advances, a knight attacks, a new business venture is launched. In complex environments, however, moves are better understood as **strategic allocations of effort or resources across multiple dimensions and options**. A single move may simultaneously improve revenue, secure assets, and build skills, each with different probabilities of success and potential risks. Moves transform positions into new positions, reshaping the distribution of potential future outcomes.

This perspective naturally leads to **mixed strategies**, not just in games but in life. Mixed strategies involve distributing resources across multiple candidate actions rather than committing entirely to one. Doing so allows an agent to **balance advantage and threat**, preserve optionality, and maximize the expected growth of their position. In Chess, controlling the center not only increases the probability of a successful attack but creates multiple ways to reach victory while limiting the opponent’s options. In business or life, diversifying efforts across complementary ventures can mitigate downside risk while expanding the potential for success.

In this article, we propose a generalized framework for reasoning about decision-making across domains. Positions are represented as multi-dimensional vectors encoding the agent’s state. Moves are treated as **strategy tensors** allocating resources across dimensions and candidate actions. Utility is not merely the expected return but a combination of **probability of success, risk of failure, and the diversity of available pathways**. Mixed strategies emerge naturally as a rational response to multi-dimensional constraints and threats, reflecting both the **likelihood and robustness of future success**.

By formalizing the relationship between an agent’s position, moves, probabilistic outcomes, and the structure of possible futures, we lay the foundation for a **universal approach to rational action in complex systems** — one that accounts for advantage and threat, growth and robustness, across multiple dimensions of strategy, even under imperfect information.

## Position, Advantage, Threat, and Path Structure
Strategic decision-making begins from an **agent’s position** $P = [p_1, p_2, \dots, p_n]$, which encodes the complete state of the agent across relevant dimensions such as resources, skills, knowledge, environment, and capabilities. This vector represents **everything the agent knows and controls that is relevant to decision-making**. The position is not the full global state of the system or the game; it is the **projection of the world onto the agent’s decision surface**, capturing what is actionable and observable.

From any position $P$, the agent faces multiple possible continuation paths that lead to future outcomes $O = {o_1, o_2, \dots, o_m}$. Each outcome occurs with probability

$$
Pr(o_i \mid P)
$$

and carries a value

$$
V(o_i)
$$

representing the desirability of that outcome from the agent’s perspective. These probabilities reflect both uncertainty from the environment and the actions of other agents.

The **advantage** of a position quantifies the expected magnitude of favorable outcomes accessible from $P$. Let

$$
O^+ = { o_i \in O : V(o_i) > 0 }
$$

denote the set of desirable outcomes. Advantage is then

$$
A(P) = \sum_{o_i \in O^+} Pr(o_i \mid P) , V(o_i)
$$

capturing how much positive value the agent can access.

Similarly, the **threat exposure** of a position measures the expected magnitude of harmful outcomes. Let

$$
O^- = { o_i \in O : V(o_i) < 0 }
$$

denote undesirable outcomes. Threat is defined as

$$
T(P) = \sum_{o_i \in O^-} Pr(o_i \mid P) , |V(o_i)|
$$

Both advantage and threat are derived from the **same probability distribution over outcomes**, ensuring consistency. Crucially, **eliminating threats does not automatically increase advantage**; it restructures the space of possible futures, improving robustness without necessarily creating new positive value.

However, expected value alone does not capture the full strategic quality of a position. Two positions with similar $A(P)$ and $T(P)$ can differ dramatically in **path diversity** — the number and structural variety of ways to reach success or failure. Consider two Chess positions where victory is likely: in one, only a single precise sequence leads to success; in the other, multiple distinct sequences achieve the same outcome. The second position is more robust and strategically valuable.

Formally, let the possible continuation paths from position $P$ be

$$
\pi_1, \pi_2, \dots, \pi_k
$$

with probabilities

$$
Pr(\pi_j \mid P)
$$

and outcomes

$$
o(\pi_j)
$$

We partition the paths according to favorable or harmful outcomes:

$$
\Pi^+ = { \pi_j : V(o(\pi_j)) > 0 }, \quad
\Pi^- = { \pi_j : V(o(\pi_j)) < 0 }
$$

The **diversity of success paths** is the entropy of the probability distribution over paths leading to positive outcomes:

$$
D^+(P) = - \sum_{\pi_j \in \Pi^+} Pr(\pi_j \mid P) \log Pr(\pi_j \mid P)
$$

Similarly, the **diversity of failure paths** measures how many distinct ways the agent could end up in undesirable outcomes:

$$
D^-(P) = - \sum_{\pi_j \in \Pi^-} Pr(\pi_j \mid P) \log Pr(\pi_j \mid P)
$$

This symmetry captures a key strategic principle: a strong position **simultaneously increases the number of viable paths to success while collapsing the number of distinct paths to failure**. Offensive moves often raise $D^+(P)$ by generating multiple avenues to favorable outcomes, while defensive moves reduce $D^-(P)$ by eliminating classes of losing continuations.

Combining these components, the **utility of a position** is:

$$
U(P) = A(P) - T(P) + D^+(P) - D^-(P)
$$
